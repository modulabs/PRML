````python
from sklearn.datasets import load_iris
````


````python
dataset = load_iris()
````

````python
print(f"{len(dataset['data'])},{len(dataset['target'])}")
````

````python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.multivariate.pca import PCA
````

# PCA with 4

````python
pc4 = PCA(dataset['data'], standardize=False, ncomp=4)
pc4.plot_scree(ncomp=4)
pc4.plot_rsquare()
plt.show()
````

# PCA with 3

````python
pc3=PCA(dataset['data'], standardize=False, ncomp=3)
pc3.plot_scree()
pc3.plot_rsquare()
plt.show()
````

# PCA with 2

````python
pc2=PCA(dataset['data'], standardize=False, ncomp=2)
pc2.plot_scree()
pc2.plot_rsquare()
plt.show()
````

# PCA with 1

````python
pc1=PCA(dataset['data'], standardize=False, ncomp=1)
pc1.plot_scree()
pc1.plot_rsquare()
plt.show()
````
