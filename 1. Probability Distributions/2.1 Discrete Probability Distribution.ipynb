### Bernoulli Distribution

### $Bern(x|\mu) = \mu^x(1-\mu)^{1-x}$



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bern(x,mu):
    return np.power(mu,x)*np.power((1-mu), (1-x))

x = np.linspace(0,1,2)
y = bern(x,0.4)
plt.bar(x,y)
plt.show()
```

### Binomial Distribution

### $Bin(m|N,\mu) = \binom{N}{m}\mu^m(1-\mu)^{N-m}$
### $\binom{N}{m} \equiv \frac{N!}{(N-m)!m!}$


```python
from scipy.special import factorial
#numpy factorial 오류나서 scipy사용

fact = factorial
pow = np.power

def bino(m,N,mu):
    return (fact(N)/(fact(N-m)*fact(m)))*pow(mu,m)*pow((1-mu),(N-m))    


N=10
mu=0.4

x= np.linspace(1,100,100)
y = bino(x,N,mu)

plt.bar(x,y)
```

### Beta Distribution

### $Beta(\mu|a,b) = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1}$


```python
from math import gamma
pow = np.power

def beta(mu,a,b):
    return (gamma(a+b)/gamma(a)*gamma(b))*pow(mu,(a-1))*pow((1-mu),(b-1))
    
```

```python
x = np.linspace(0,1,10000)
y = beta(x,0.8,0.8)
plt.scatter(x,y,s=0.5)
plt.show()

x = np.linspace(0,1,10000)
y = beta(x,1,1)
plt.scatter(x,y,s=0.5)
plt.show()

x = np.linspace(0,1,10000)
y = beta(x,2,3)
plt.scatter(x,y,s=0.5)
plt.show()

x = np.linspace(0,1,10000)
y = beta(x,8,4)
plt.scatter(x,y,s=0.5)
plt.show()
```

### $p(\mu|m,l,a,b) = \frac{\Gamma(m+a+l+b)}{\Gamma(m+a)\Gamma(l+b)}\mu^{m+a-1}(1-\mu)^{l+b-1}$


```python
from math import gamma
pow = np.power

def beta_p(mu,m,l,a,b):
    return (gamma(m+a+l+b)/gamma(m+a)*gamma(l+b))*pow(mu,(m+a-1))*pow((1-mu),(1+b-1))
```

```python
#사전분포
x = np.linspace(0,1,10000)
y = beta_p(x,1,0,2,2)
plt.scatter(x,y,s=0.5)
plt.show()

#가능도함수
x = np.linspace(0,1,1000)
y = np.linspace(0,1,1000)
plt.ylim(0,2)
plt.scatter(x,y,s=0.5)
plt.show()

#사후분포
x = np.linspace(0,1,10000)
y = beta_p(x,1,0,3,2)
plt.scatter(x,y,s=0.5)
plt.show()
```
